# Problem 2 : Type for ‚Äúfood‚Äù in search

Type for ‚Äúfood‚Äù in search bar and print all the names of the Instagram Handles that are displayed in list after typing ‚Äúfood‚Äù bar and print all the names of the Instagram Handles that are displayed in list after typing ‚Äúfood‚Äù
Note : Make sure to avoid printing hashtags


```python
from selenium.common.exceptions import TimeoutException

def Search(string):
    try:
        # 1. Click the Search Icon
        wait.until(EC.element_to_be_clickable((
            By.XPATH, "/html/body/div[1]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div/div/div/div/div[2]/div[2]/span/div/a/div"
        ))).click()
        time.sleep(2)

        # 2. Type in the search box
        search_box = wait.until(EC.presence_of_element_located((
            By.XPATH, "//input[@placeholder='Search']"
        )))
        search_box.send_keys(string)
        time.sleep(3)

        # 3. Use XPath based on your given structure
        username_spans = wait.until(EC.presence_of_all_elements_located((
            By.XPATH, "//div[contains(@class, 'x1y1aw1k')]//a/div[1]/div/div/div[2]/div/div/div/span"
        )))

        usernames = []
        for span in username_spans:
            text = span.text.strip()
            if text and not text.startswith("#") and text not in usernames:
                usernames.append(text)

        # 4. Print result
        print(f"\nSearch results for '{string}':")
        for name in usernames:
            print(name)
            
        time.sleep(1)
        # 5. Clear search
        clear_button = wait.until(EC.element_to_be_clickable((
            By.XPATH, "//div[@role='button' and @aria-label='Clear the search box']"
        )))
        clear_button.click()
        time.sleep(2)
        wait.until(EC.element_to_be_clickable((
            By.XPATH, "/html/body/div[1]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div/div/div/div/div[2]/div[2]/span/div/a/div"
        ))).click()

    except TimeoutException:
        print("Something Went Wrong!")
```

# Problem 5 : Like/Unlike posts
Liking the top 30 posts of the ‚Äòdilsefoodie'. Print message if you have already liked it.
Unliking the top 30 posts of the ‚Äòdilsefoodie‚Äô. Print message if you have already unliked it.

```python

def liking_page_post():
    try:
        wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "_aagw"))).click()
        print("‚úÖ Opened the first post.")
        time.sleep(2) 

    except TimeoutException:
        print("‚ùå Could not find the first post to click. Stopping.")
        return

    for i in range(30):
        print(f"\n--- Processing Post #{i+1} ---")
        
        is_already_liked = False
        try:
            WebDriverWait(driver, 2).until(EC.presence_of_element_located((
                By.XPATH, "//article//*[local-name()='svg' and @aria-label='Unlike']"
            )))
            print("üëç Post is already liked.")
            is_already_liked = True

        except TimeoutException:
            print("Post not liked yet, attempting to like...")
            is_already_liked = False

        if not is_already_liked:
            try:
                # We will still locate the button in the same way
                like_button = wait.until(EC.presence_of_element_located((
                    By.XPATH, "(//article//section//div[@role='button'])[1]"
                )))
                
                # === THE FIX: USE A JAVASCRIPT CLICK ===
                # Instead of like_button.click(), we execute a script.
                # This is more powerful and can bypass issues where a normal click doesn't register.
                driver.execute_script("arguments[0].click();", like_button)
                
                print("‚ù§Ô∏è Liked the post!")
                time.sleep(2)
            except TimeoutException:
                print("‚ùå Could not find the like button to click. The layout may have changed again.")

        # Click the "Next" button 
        try:
            next_button = wait.until(EC.element_to_be_clickable((
                By.XPATH, "//*[local-name()='svg' and @aria-label='Next']/ancestor::div[contains(@class, '_aaqg')]"
            )))
            next_button.click()
            print("‚û°Ô∏è Clicked 'Next'.")
            time.sleep(3) 

        except TimeoutException:
            print("üèÅ No more posts found. Stopping the process.")
            break

    # Close the post modal
    try:
        close_button = wait.until(EC.element_to_be_clickable((
            By.XPATH, "//*[local-name()='svg' and @aria-label='Close']/ancestor::div[@role='button']"
        )))
        close_button.click()
        print("‚úÖ Closed post view.")
    except TimeoutException:
        print("Could not find close button, navigating back instead.")
        driver.back()
```

```python
def unliking_page_post():
    try:
        # Scroll down to reveal the first post
        driver.execute_script("window.scrollTo(0, 500);")
        time.sleep(2)

        # Close any popup/dropdown if visible
        try:
            close_btns = driver.find_elements(By.XPATH, "//div[@role='dialog']//button")
            for btn in close_btns:
                if btn.is_displayed():
                    btn.click()
                    print("üßπ Closed a popup.")
                    time.sleep(1)
        except:
            pass

        # Locate first post using updated path and click
        post = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "_aagw")))
        driver.execute_script("arguments[0].scrollIntoView(true);", post)
        driver.execute_script("arguments[0].click();", post)
        print("‚úÖ Opened the first post.")
        time.sleep(2)

    except Exception as e:
        print(f"‚ùå Could not open the first post: {e}")
        return

    for i in range(30):
        print(f"\n--- Processing Post #{i+1} ---")
        print("üì∏ Post URL:", driver.current_url)
        try:
            unlike_btn = WebDriverWait(driver, 2).until(EC.presence_of_element_located((
                By.XPATH, "//section//span//*[local-name()='svg' and @aria-label='Unlike']/ancestor::div[@role='button']"
            )))
            driver.execute_script("arguments[0].click();", unlike_btn)
            print("üíî Unliked the post.")
            time.sleep(2)
        except TimeoutException:
            print("üëç Post already unliked or like button not found.")
        except Exception as e:
            print(f"‚ö†Ô∏è Error while unliking: {e}")

        # Try multiple XPaths for the "Next" button
        try:
            next_btn = None
            next_xpaths = [
                "//*[local-name()='svg' and @aria-label='Next']/ancestor::div[@role='button']",
                "//button//*[local-name()='svg' and @aria-label='Next']/ancestor::button"
            ]
            for path in next_xpaths:
                try:
                    next_btn = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.XPATH, path)))
                    break
                except:
                    continue

            if next_btn:
                driver.execute_script("arguments[0].click();", next_btn)
                print("‚û°Ô∏è Moved to next post.")
                time.sleep(2)
            else:
                print("üèÅ No Next button found. Exiting.")
                break

        except Exception as e:
            print(f"üèÅ Reached end or next failed: {e}")
            break

    # Close modal
    try:
        close_btn = wait.until(EC.element_to_be_clickable((
            By.XPATH, "//*[local-name()='svg' and @aria-label='Close']/ancestor::div[@role='button']"
        )))
        driver.execute_script("arguments[0].click();", close_btn)
        print("‚úÖ Closed the post view.")
    except:
        print("‚ö†Ô∏è Could not close the modal. Trying to go back.")
        driver.back()
```

```python
def Extracting_list_of_avaliable_followers():
    """
    Navigates to the followers list of a profile, extracts visible usernames,
    and then returns them. This function will now leave the modal open,
    giving the user control over when to close it.
    Assumes 'driver' and 'wait' are globally available.
    """
    try:
        # Locating the Followers button and click on it
        followers_button = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div/div/div[2]/div/div/div[1]/div[2]/div[1]/section/main/div/header/section[3]/ul/li[2]/div/a')))
        followers_button.click()
        print("‚úÖ Clicked on Followers button.")
        # Increased sleep time to give the modal more time to load and for user to observe
        time.sleep(5) 

        All_User = set() # Use a set to store unique usernames to avoid duplicates

        # Locate all user handles that are currently visible in the modal.
        handles = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//div[@role="dialog"]//div[contains(@class, "x9f619")]//span/div/a/div/div/span')))
        
        for h in handles:
            username = h.text.split()[0] # Extract the username (assuming no spaces)
            if username: # Ensure username is not empty before adding
                All_User.add(username)
        
        print(f"‚úÖ Extracted {len(All_User)} unique user handles from the initially visible list.")
        print("Note: If the platform no longer supports dynamic scrolling for followers, only the initially visible users can be extracted.")

        # This function no longer closes the modal.
        # It is now the responsibility of the calling script to close the modal
        # after the desired actions (like printing) have been completed.
        
        # Convert set to list for consistent return type
        return list(All_User)

    except TimeoutException:
        print("‚ùå Timeout: Something went wrong while waiting for elements (e.g., followers button or list).")
        return [] # Return empty list on timeout
    except Exception as e:
        print(f"‚ùå An unexpected error occurred: {e}")
        return [] # Return empty list on other errors


```

```python

def Extracting_list_of_user_following():
    """
    Navigates to the following list of a profile, scrolls to the end,
    extracts all usernames, and returns a list.
    This function implements a corrected version of the user's logic.
    Assumes 'driver' and 'wait' are globally available.
    """
    try:
        # Step 1: Click "Following"
        wait.until(EC.element_to_be_clickable((By.XPATH, '//section/main//li[3]//a'))).click()
        print("‚úÖ Clicked on Following button.")
        time.sleep(2)

        # Step 2: Scroll the following modal using the provided XPath
        scroll_box = wait.until(EC.presence_of_element_located(
            (By.XPATH, '/html/body/div[4]/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[3]')
        ))
        print("‚úÖ Identified scrollable element within the modal.")
        
        last_height = 0
        all_users = set()
        
        # Correctly implementing a scrolling and extraction loop
        while True:
            # Find all username elements currently visible in the scroll box
            username_elements = scroll_box.find_elements(By.XPATH, './/div/div/div/div/div[2]/div/div/div/div/span/div/a/div/div/span')
            for elem in username_elements:
                username = elem.text.strip()
                if username:
                    all_users.add(username)

            driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scroll_box)
            time.sleep(1.5)
            new_height = driver.execute_script("return arguments[0].scrollHeight", scroll_box)

            if new_height == last_height:
                print("No new content loaded. Reached end of list.")
                break
            
            last_height = new_height

        print("‚úÖ Finished scrolling.")
        print(f"‚úÖ Total users extracted: {len(all_users)}")

        return list(all_users)

    except TimeoutException:
        print("‚ùå Timeout: List box not found.")
        return []
    except Exception as e:
        print(f"‚ùå Error: {e}")
        return []
```
